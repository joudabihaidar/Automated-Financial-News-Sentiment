```python
import re

def extract_ticker_symbol(url):
    match = re.search(r'/([^/]+)/?$', url)
    if match:
        return match.group(1)
    else:
        return None

# Example usage
urls = [
    "https://finance.yahoo.com/quote/AAPL/news",
    "https://finance.yahoo.com/quote/AAPL/m/e5b1f221-8500-38e4-97a2-ebdf6a335ea1/is-meta-stock-a-buy-after.html"
]

for url in urls:
    symbol = extract_ticker_symbol(url)
    print(f"URL: {url}, Ticker Symbol: {symbol}")
```
